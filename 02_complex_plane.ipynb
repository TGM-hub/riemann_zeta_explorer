# 02 — ζ(s) dans le plan complexe

```python
import numpy as np
import mpmath
import plotly.graph_objects as go
from plotly.subplots import make_subplots

mpmath.mp.dps = 25

TEMPLATE     = "plotly_dark"
COLOR_MAIN   = "#7C6FCD"
COLOR_ACCENT = "#F2A623"
COLOR_MUTED  = "#888780"
COLOR_DANGER = "#E24B4A"
COLOR_TEAL   = "#1D9E75"
```

## ζ(s) pour s complexe — les termes tournent

Quand $s = \sigma + it$ est complexe, chaque terme de la série devient :

$$\frac{1}{n^s} = \frac{1}{n^{\sigma + it}} = \frac{1}{n^\sigma} \cdot e^{-it \ln n}$$

Le facteur $e^{-it\ln n} = \cos(t\ln n) - i\sin(t\ln n)$ est un vecteur qui **tourne** dans le plan complexe.
Chaque terme $n$ tourne à une vitesse différente ($\ln n$), avec une amplitude $\frac{1}{n^\sigma}$ qui décroît.

La somme $\zeta(s)$ est donc une addition de vecteurs tournants. Elle trace des **spirales** quand $t$ varie.

```python
# Trajectoire de ζ(2+it) dans le plan complexe
t_vals = np.linspace(0, 50, 2000)

zeta_s2   = [complex(mpmath.zeta(2.0 + 1j*t)) for t in t_vals]
re_s2     = [z.real for z in zeta_s2]
im_s2     = [z.imag for z in zeta_s2]

fig = go.Figure()
fig.add_trace(go.Scatter(
    x=re_s2, y=im_s2,
    mode="lines",
    line=dict(color=COLOR_MAIN, width=1.5),
    hovertemplate="Re=%{x:.3f}<br>Im=%{y:.3f}<extra>σ=2</extra>"
))

# Point de départ : ζ(2) = π²/6
fig.add_trace(go.Scatter(
    x=[np.pi**2/6], y=[0],
    mode="markers",
    marker=dict(color=COLOR_ACCENT, size=10),
    name="ζ(2) = π²/6  (t=0)"
))

fig.add_trace(go.Scatter(
    x=[0], y=[0], mode="markers",
    marker=dict(color=COLOR_DANGER, size=10, symbol="circle-open", line_width=2),
    name="Origine (0,0)"
))

fig.update_layout(
    template=TEMPLATE,
    title="Trajectoire de ζ(2+it) dans ℂ — t ∈ [0, 50]",
    xaxis_title="Re(ζ)", yaxis_title="Im(ζ)",
    height=480
)
fig.show()
print(f"t=0  : ζ(2+0i) = {zeta_s2[0]:.4f}  (= π²/6 = {np.pi**2/6:.4f})")
print(f"La trajectoire boucle mais ne passe jamais par l'origine pour σ=2.")
```

## Est-ce que ζ peut s'annuler ?

Pour $\sigma = 2$, la trajectoire boucle mais ne passe pas par l'origine.  
Qu'en est-il sur la **ligne critique** $\sigma = \frac{1}{2}$ ?

```python
# Trajectoire de ζ(½+it) — la ligne critique
zeta_crit = [complex(mpmath.zeta(0.5 + 1j*t)) for t in t_vals]
re_c      = [z.real for z in zeta_crit]
im_c      = [z.imag for z in zeta_crit]

fig = go.Figure()
fig.add_trace(go.Scatter(
    x=re_c, y=im_c,
    mode="lines",
    line=dict(color=COLOR_DANGER, width=1.5),
    hovertemplate="Re=%{x:.3f}<br>Im=%{y:.3f}<extra>σ=½</extra>"
))
fig.add_trace(go.Scatter(
    x=[0], y=[0], mode="markers",
    marker=dict(color=COLOR_ACCENT, size=14, symbol="circle"),
    name="Origine — zéros ici"
))
fig.update_layout(
    template=TEMPLATE,
    title="Trajectoire de ζ(½+it) dans ℂ — t ∈ [0, 50]",
    xaxis_title="Re(ζ)", yaxis_title="Im(ζ)",
    height=480
)
fig.show()
print("La trajectoire passe par l'origine → ce sont les zéros non triviaux.")
```

## Les zéros non triviaux — vérification numérique

Un **zéro non trivial** de ζ est un $s = \frac{1}{2} + it$ tel que $\zeta(s) = 0$ exactement.  
Le premier est à $t \approx 14.134725$.

```python
# Vérification du premier zéro
t_zero = 14.134725141734693
val = complex(mpmath.zeta(0.5 + 1j * t_zero))
print(f"ζ(½ + {t_zero}i) = {val.real:.2e} + {val.imag:.2e}i")
print(f"|ζ| = {abs(val):.2e}  ← pratiquement zéro")

# Les 10 premiers zéros non triviaux connus
first_zeros = [14.1347, 21.0220, 25.0109, 30.4249, 32.9351,
               37.5862, 40.9187, 43.3271, 48.0052, 49.7738]

print("\n|ζ(½+it)| aux 10 premiers zéros connus :")
for i, t in enumerate(first_zeros, 1):
    mod = abs(complex(mpmath.zeta(0.5 + 1j*t)))
    print(f"  ρ_{i:2d} : t={t:.4f}   |ζ| = {mod:.2e}")
```

## Les zéros sont-ils vraiment sur Re(s) = ½ ?

On teste numériquement : pour $\sigma = 0.4$, $0.5$, $0.6$ — est-ce que $|\zeta(\sigma+it)|$ s'annule ?

Si l'hypothèse de Riemann est vraie, seul $\sigma = \frac{1}{2}$ produit de vrais zéros.

```python
sigmas = [0.4, 0.5, 0.6]
colors = [COLOR_MAIN, COLOR_DANGER, COLOR_TEAL]

fig = go.Figure()

for sigma, color in zip(sigmas, colors):
    mods = [abs(complex(mpmath.zeta(sigma + 1j*t))) for t in t_vals]
    fig.add_trace(go.Scatter(
        x=t_vals, y=mods,
        name=f"σ = {sigma}",
        line=dict(color=color, width=1.5),
        hovertemplate=f"t=%{{x:.2f}}<br>|ζ|=%{{y:.4f}}<extra>σ={sigma}</extra>"
    ))
    print(f"σ={sigma} — min|ζ| = {min(mods):.6f}")

# Marqueurs des zéros connus sur σ=0.5
fig.add_trace(go.Scatter(
    x=first_zeros, y=[0]*len(first_zeros),
    mode="markers",
    marker=dict(color=COLOR_ACCENT, size=9, symbol="circle"),
    name="Zéros non triviaux (σ=½)"
))

fig.add_hline(y=0, line_color=COLOR_MUTED, opacity=0.3, line_width=0.8)

fig.update_layout(
    template=TEMPLATE,
    title="|ζ(σ+it)| — seul σ=½ s'annule vraiment",
    xaxis_title="t = Im(s)",
    yaxis_title="|ζ(σ+it)|",
    height=450,
    hovermode="x unified"
)
fig.show()
```

## Zéros triviaux vs zéros non triviaux

Il existe deux types de zéros de ζ :

| Type | Localisation | Explication |
|---|---|---|
| **Triviaux** | $s = -2, -4, -6, \ldots$ | Forcés par l'équation fonctionnelle : $\sin(\pi s/2) = 0$ |
| **Non triviaux** | $s = \frac{1}{2} + it$ (conjecturé) | Mystérieux — aucune formule fermée |

Les zéros triviaux sont **explicables** — ils viennent d'un facteur $\sin$ dans la formule de prolongement analytique.  
Les zéros non triviaux, eux, sont **imprévisibles** — on ne sait pas les calculer sans chercher numériquement.

**L'hypothèse de Riemann** : tous les zéros non triviaux ont $\text{Re}(s) = \frac{1}{2}$.

```python
# Visualiser les deux types de zéros dans le plan complexe (s = σ + it)
fig = go.Figure()

# Zéros triviaux — sur l'axe réel négatif
trivial = [-2, -4, -6, -8, -10]
fig.add_trace(go.Scatter(
    x=trivial, y=[0]*len(trivial),
    mode="markers",
    marker=dict(color=COLOR_MUTED, size=10, symbol="circle-open", line_width=2),
    name="Zéros triviaux  (−2, −4, −6...)"
))

# Zéros non triviaux — sur la ligne critique
nontrivial_t = first_zeros + [-t for t in first_zeros]  # symétrie conjuguée
fig.add_trace(go.Scatter(
    x=[0.5]*len(nontrivial_t), y=nontrivial_t,
    mode="markers",
    marker=dict(color=COLOR_DANGER, size=9, symbol="circle"),
    name="Zéros non triviaux  (Re=½)"
))

# Ligne critique
fig.add_vline(x=0.5, line_dash="dash", line_color=COLOR_DANGER, opacity=0.4,
              annotation_text="Re(s) = ½", annotation_font_color=COLOR_DANGER)

# Pôle en s=1
fig.add_trace(go.Scatter(
    x=[1], y=[0], mode="markers",
    marker=dict(color=COLOR_ACCENT, size=13, symbol="x"),
    name="Pôle (s=1)"
))

fig.update_layout(
    template=TEMPLATE,
    title="Zéros de ζ dans le plan complexe — triviaux vs non triviaux",
    xaxis_title="σ = Re(s)",
    yaxis_title="t = Im(s)",
    xaxis_range=[-11, 3],
    height=520
)
fig.show()
```

## Ce qu'on a compris

| Observation | Ce que ça signifie |
|---|---|
| Les termes $1/n^s$ tournent dans ℂ | $e^{it\ln n}$ — chacun à sa vitesse |
| ζ(s) trace des spirales quand $t$ varie | Addition de vecteurs tournants |
| La trajectoire passe par l'origine sur $\sigma=½$ | → zéros non triviaux |
| $\sigma=0.4$ et $\sigma=0.6$ ne s'annulent pas | Les zéros semblent bien sur $\sigma=½$ |
| Zéros triviaux en $s=-2,-4,...$ | Explicables par $\sin(\pi s/2)=0$ |
| **Hypothèse de Riemann** | Tous les zéros non triviaux sur $\sigma=½$ — non prouvé |
