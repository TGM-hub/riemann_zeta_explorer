```python
import numpy as np
import mpmath
import plotly.graph_objects as go
from plotly.subplots import make_subplots

mpmath.mp.dps = 15

TEMPLATE     = "plotly_dark"
COLOR_MAIN   = "#7C6FCD"
COLOR_ACCENT = "#F2A623"
COLOR_MUTED  = "#888780"
```

## 1. La trajectoire de ζ(½+it) dans ℂ

On fixe Re(s) = ½ et on fait varier t. La fonction ζ(½+it) trace une courbe dans le plan complexe.  
Chaque passage par l'origine **(0, 0)** correspond à un **zéro non trivial**.

Ce qu'on observe en faisant grandir t :
- La trajectoire **boucle** — les termes $e^{it\ln n}$ tournent dans tous les sens
- Elle **se densifie** — les zéros deviennent de plus en plus fréquents
- Elle finit par **remplir l'espace** — apparent chaos, mais structure sous-jacente

On visualise trois intervalles côte à côte pour voir cette évolution.

```python
ranges = [(0.1, 50), (50, 200), (200, 500)]
titles = ["t ∈ [0, 50]", "t ∈ [50, 200]", "t ∈ [200, 500]"]

fig = make_subplots(rows=1, cols=3, subplot_titles=titles)

for col, (t_start, t_end) in enumerate(ranges, 1):
    print(f"Calcul t ∈ [{t_start}, {t_end}]...")
    t_sub  = np.linspace(t_start, t_end, 5000)
    z_sub  = [complex(mpmath.zeta(0.5 + 1j*t)) for t in t_sub]
    re_sub = [z.real for z in z_sub]
    im_sub = [z.imag for z in z_sub]

    fig.add_trace(go.Scatter(
        x=re_sub, y=im_sub,
        mode="markers",
        marker=dict(color=t_sub, colorscale="Viridis", size=1.5),
        showlegend=False,
        hovertemplate="Re=%{x:.3f}<br>Im=%{y:.3f}<extra></extra>"
    ), row=1, col=col)

    # Origine
    fig.add_trace(go.Scatter(
        x=[0], y=[0], mode="markers",
        marker=dict(color=COLOR_ACCENT, size=10, symbol="circle"),
        showlegend=(col == 1),
        name="Origine — zéros ici"
    ), row=1, col=col)

fig.update_layout(
    template=TEMPLATE, height=480,
    title="Trajectoire de ζ(½+it) — la courbe se densifie avec t"
)
fig.update_xaxes(title_text="Re(ζ)")
fig.update_yaxes(title_text="Im(ζ)", col=1)

fig.show()
print("Plus t est grand, plus la trajectoire remplit l'espace.")
print("Les zéros deviennent de plus en plus fréquents — densité ∝ ln(t)/2π.")
```

## 2. Animation — la trajectoire se trace en temps réel

L'animation montre ζ(½+it) se dessiner progressivement dans le plan complexe :
- Un point rouge suit la trajectoire, **t** s'affiche en temps réel
- Le point **ralentit et passe au jaune** à l'approche d'un zéro non trivial
- Les zéros trouvés s'accumulent sur la droite : ρ₁: t=14.1347, ρ₂: t=21.0220...
